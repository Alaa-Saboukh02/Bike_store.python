### 3] Data Wrangling (Cleaning & Manipulation)

##### Are there any missing / duplicated data?

```python
print(Bike_store.isnull().sum())
```

```python
print(Bike_store.duplicated())
```

```python
print("Missing data per column:")
print(Bike_store.isnull().sum())

print("\nNumber of duplicated rows:")
print(Bike_store.duplicated().sum())
```

```python
for column in Bike_store.columns:
    if Bike_store[column].dtype == 'object': 
        Bike_store[column].fillna(Bike_store[column].mode()[0], inplace=True)
    else:  
        Bike_store[column].fillna(Bike_store[column].mean(), inplace=True)

print(Bike_store)
```

```python
Bike_store_cleaned = Bike_store.dropna()
print(Bike_store_cleaned)
```

```python
#Sort Values
Bike_store_sorted = Bike_store.sort_values(by='order_date')
print(Bike_store_sorted)
```

```python
Bike_store_sorted = Bike_store.sort_values(by=['revenue', 'order_id'], ascending=[True, False])
print(Bike_store_sorted)
```

```python
#Add Column
Bike_store['Total revenue'] = Bike_store['revenue'] + Bike_store['total_units']
```

### 4] Data Analysis & Visualization (EDA & Statistical Analysis)
#### Exploratory Data Analysis

##### What is the average Revenue?

```python
manual_average = sum(Bike_store['REVENUE']) / len(Bike_store['REVENUE'])
print(manual_average)
```

##### What were the highest and lowest Revenue?

```python
highest_age = Bike_store['REVENUE'].max()
lowest_age = Bike_store['REVENUE'].min()

print(f"Highest revenue: {highest_age}")
print(f"Lowest revenue: {lowest_age}")
```

##### How many orders have 'Road Bikes'  as a category name ?

```python
road_bikes_orders = Bike_store[Bike_store['CATEGORY_NAME'] == 'Road Bikes']
```

```python
road_bikes_orders
```

```python
num_road_bikes_orders = road_bikes_orders.shape[0]
print(num_road_bikes_orders)
```

##### Display the data which revenue greater than 10000

```python
high_revenue_orders = Bike_store[Bike_store['REVENUE'] > 10000]
high_revenue_orders
```

##### what is the total revenue for each Brand ?

```python
total_revenue_by_brand =Bike_store.groupby('BRAND_NAME')['REVENUE'].sum().reset_index()
total_revenue_by_brand
```

##### what is the total revenue for each Category Name ?

```python
total_revenue_by_Category =Bike_store.groupby('CATEGORY_NAME')['REVENUE'].sum().reset_index()
total_revenue_by_Category
```
